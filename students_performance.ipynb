# EDA

1. Printing unique values of each column using unique()
2. Calcualation of total score and average score to determine the performance in each subject

```python
df.columns
```

```python
#Check categories in each columns
print("Categories in 'gender' variable: ", end=" ")
print(df['gender'].unique())

print("Categories in 'race/ethnicity' variable: ", end=" ")
print(df['race/ethnicity'].unique())

print("Categories in 'parental level of education' variable: ", end=" ")
print(df['parental level of education'].unique())

print("Categories in 'lunch' variable: ", end=" ")
print(df['lunch'].unique())

print("Categories in 'test preparation course' variable: ", end=" ")
print(df['test preparation course'].unique())
```

```python
#Adding columns for Total score and Average score
df['Total Score'] = df['math score'] + df['reading score'] + df['writing score']
df['Average Score'] = df['Total Score'] / 3
df.head()
```

```python
# Count the number of students with full marks (100) in each subject
reading_full = df[df['reading score'] == 100]['Average Score'].count()
writing_full = df[df['writing score'] == 100]['Average Score'].count()
math_full = df[df['math score'] == 100]['Average Score'].count()

# Print the count of students with full marks in each subject
print('Number of students with full marks in Maths: ', math_full)
print('Number of students with full marks in Writing: ', writing_full)
print('Number of students with full marks in Reading: ', reading_full)
```

```python
# Count the number of students with less than or equal to 20 marks in each subject
reading_less_20 = df[df['reading score'] <= 20]['Average Score'].count()
writing_less_20 = df[df['writing score'] <= 20]['Average Score'].count()
math_less_20 = df[df['math score'] <= 20]['Average Score'].count()

# Print the count of students with less than or equal to 20 marks in each subject
print('Number of students with less than or equal to 20 marks in Maths: ',math_less_20)
print('Number of students with less than or equal to 20 marks in Writing: ',writing_less_20)
print('Number of students with less than or equal to 20 marks in Reading: ', reading_less_20)
```

**Observation**:<br>
Students perform better in reading as compared to the other subjects<br>
Maths has the worst performance amongst the 3 subjects

# DATA VISUALIZATION

```python
#subplots(row, column) #figsize(width, height)
fig, axs = plt.subplots(1, 2, figsize=(15, 7)) 
fig.add_subplot(121)
sns.histplot(data=df,x='Average Score',bins=25,color='b')
plt.title('Distribution of Average Scores')
plt.xlabel('Average Score')
plt.ylabel('Frequency')


plt.subplot(122)
sns.histplot(data=df,x='Average Score',kde=True,hue='gender')
plt.title('Distribution of Average Scores by Gender')
plt.xlabel('Average Score')
plt.ylabel('Frequency')
plt.show()
#plt.subplot(122) same as fig.add_subplot(122)
#kde= kernel density. This adds a curve representing kernel density estimate of the data (Visually pleasant 
#representation of data distribution)

#hue is used to determine which column in the data frame should be used for colour encoding.
```

**Observation:** Female students perform better than male students

```python
sns.histplot(data=df, x='Average Score', kde=True, hue='lunch')
plt.title('Distribution of Average Scores by Lunch (Overall)')
plt.xlabel('Average Score')
plt.ylabel('Frequency')
```

**Observation:** Students perform better when they have a standard lunch (based on Average Score)

```python
plt.figure(figsize=(12,6))
plt.subplot(2,2,1)
sns.barplot (y=df['math score'], hue=df['lunch'])
plt.subplot(2,2,2)
sns.barplot (y=df['reading score'], hue=df['lunch'])
plt.subplot(2,2,3)
sns.barplot (y=df['writing score'], hue=df['lunch'])
```

**Observation:**: Student performance is consistently good across all subjects when they have a standard lunch

```python
size = df['parental level of education'].value_counts()
labels = 'Some College', "Associate's Degree", 'High School', 'Some High School', "Bachelor's Degree", "Master's Degree"
color = ['red', 'yellow', 'green', 'blue', 'orange', 'grey']

plt.pie(size, colors=color, labels=labels, autopct='%.2f%%', shadow=True, explode=[0.1, 0, 0, 0, 0, 0])
#autopct is for writing the percentage of values
plt.title('Parental Education', fontsize=20)
plt.axis('off')
```

**Obesrvation:** Most of the parents have attended some college or have an Associate's degree while a very few have completed their Master's degree

```python
size = df['test preparation course'].value_counts()
x=['None', 'Completed']
plt.title("Test Preparation Course")
plt.ylabel("Number of Students")
plt.xlabel("Course Completion")
plt.bar(x,size)
```

**Observation:** Most of the students have not enrolled in the Test Preparation Course
